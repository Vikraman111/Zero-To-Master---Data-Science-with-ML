## OO Method  vs pyplot
* When plotting singular stuff use pyplot
* Advanced plotting use OO method

```python
over_50 = heart_disease[heart_disease["age"]>50];
over_50
```

```python
# PYPLOT method
over_50.plot.scatter(x="age",y="chol",c="target",figsize=(10,6))
```

```python
# OO Method w/ pyplot
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax,
             )
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax,
             )
ax.set_xlim([45,80])
```

```python
# OO Method
fig, ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])

#Customizing Plot
ax.set(title="Heart Disease and Cholestrol Levels",xlabel = "Age",ylabel="Cholestrol")

#Creation of Legend
ax.legend(*scatter.legend_elements(),title="Target")

#Minimum Bar
ax.axhline(y=200) # or can also be given as
ax.axhline(y=heart_disease["chol"].mean(),c="r",linestyle="--")
```

```python
 #Subplots of chol, age, talach
fig, (ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10))

#adding data to ax 0
scatter = ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])

#customizing 
ax0.set(title="Heart Disease and Cholestrol Levels",xlabel= "Age",ylabel = "Cholestrol")
ax0.legend(*scatter.legend_elements(),title = "Target")

#Adding a mean line
ax0.axhline(over_50["chol"].mean(),c="red",linestyle="--")


#thalach column
scatter= ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
#customizing
ax1.set(title="Heart Disease and Thalach Analysis", xlabel = "Age", ylabel = "Thalach")
#setting Legend
ax1.legend(*scatter.legend_elements(),title = "Target")
#adding a mean line
ax1.axhline(over_50["thalach"].mean(),c="b", linestyle="--")


#Giving both the plots together a title
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
